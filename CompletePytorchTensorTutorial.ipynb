```python
import torch

x = torch.tensor([1, 2, 3])
y = torch.tensor([9, 8, 7])

#Addition
z1 = torch.empty(3)
torch.add(x, y, out=z1)
print(z1)

z = x - y

z = torch.true_divide(x, y)

#t.add_(x)

#exponentiation
z = x.pow(2)
z = x ** 2

# simple comparison
z = x > 0
z = x < 0

# Matrix multiplication
x1 = torch.rand((2, 5))
x2 = torch.rand((5, 3))

x3 = torch.mm(x1, x2)
print(x3)

# matrix exponentiation
matrix_exp = torch.rand(5, 5)
print(matrix_exp.matrix_power(3))


# element wise mult
z = x * y
print(z)

#dot product
z = torch.dot(x, y)
print(z)

#batch matrix multiplication
batch = 32
n = 10
m = 20
p = 30

tensor1 = torch.rand((batch, n, m))
tensor2 = torch.rand((batch, m, p))

out_bmm = torch.bmm(tensor1, tensor2)
print(out_bmm.shape)

# example of broadcasting
x1 = torch.rand((5, 5))
x2 = torch.rand((1, 5))

z = x1 - x2

# other userful tensor operation
sum_x = torch.sum(x, dim=0)

values, indices = torch.max(x, dim=0) # x.max(dim=0)
values, indices = torch.min(x, dim=0)

z = torch.argmax(x, dim=0) #只返回索引
z = torch.argmin(x, dim=0)

mean_x = torch.mean(x.float(), dim=0)

z = torch.eq(x, y)
print(z)

sorted_y, indices = torch.sort(y, dim=0, descending=False)
print(sorted_y)

z = torch.clamp(x, min=0, max=10)

x = torch.tensor([1, 0, 1,1, 1], dtype=torch.bool)
print(torch.any(x))
print(torch.all(x))
```

```python
torch.manual_seed(1) #设置随机数的种子，设置之后，每次生成的随机数会是固定的
indices = torch.randperm(10)
print(indices)
```

```python
# tensor indexing
import torch

batch_size = 10
features = 25

x = torch.rand((batch_size, features))

print(x[0]) #get first example
print(x[0].shape)

print(x[0, :]) # get first example
print(x[0, :].shape)


print(x[:, 0]) # get first feature of all examples
print(x[:, 0].shape)

#get the first ten features of the third examples
print(x[2, 0:10])


# fancy indexing
x = torch.arange(10)
indices = [2, 5, 8]

print(x[indices])

x = torch.rand((3, 5))
rows = torch.tensor([1, 0])
cols = torch.tensor([4, 0])

print(x[rows, cols]) # pick out x(1, 4) and x(0, 0)

#more advanced indexing
x = torch.arange(10)

print(x[(x < 2) | (x > 8)])

print(x[x.remainder(2) == 0]) # pick out all the even elements

#useful operations
print(torch.where(x > 5, x, x*2)) # if x > 5 ,set x to x, else set x to x*2
print(torch.tensor([0,0,1,2,2,3,4]).unique())

print(x.ndimension())
print(x.numel()) # count the elements in tensor
```

```python
#tensor reshaping
x = torch.arange(9)

x_3_3 = x.view(3,3) #continuous memory
print(x_3_3.shape)
print(x_3_3)

x_3_3_reshape = x.reshape(3, 3)
print(x_3_3_reshape.shape)

y = x_3_3.t() #tensor transpose
print(y)

print(y.contiguous().view(9))
#print(y.view(9)) #RuntimeError: view size is not compatible with input tensor's size and stride (at least one dimension spans across two contiguous subspaces). Use .reshape(...) instead.

x1 = torch.rand((2, 5))
x2 = torch.rand((2, 5))

print(torch.cat((x1, x2), dim=0).shape)

print(torch.cat((x1, x2), dim=1).shape)

z = x1.view(-1) #flattern x1
print(z.shape)

batch = 64
x = torch.rand((batch, 2, 5))
z = x.view(batch, -1)
print(z.shape)

#switch axis
z = x.permute(0,2, 1)
print(z.shape)

#unsqueeze
x = torch.arange(10)
print(x.unsqueeze(0).shape)
print(x.unsqueeze(1).shape)

x = torch.arange(10).unsqueeze(0).unsqueeze(1)
print(x.shape)

z = x.squeeze(1)
print(z.shape)
```
